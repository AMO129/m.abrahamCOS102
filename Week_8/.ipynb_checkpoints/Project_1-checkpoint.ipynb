```python
# We import tkinter so we can build a window-based GUI
import tkinter as tk
from tkinter import messagebox


# PARENT CLASS
class ZenithBank:

    def __init__(self, employee_name):
        # EMPLOYEE NAMES 
        self.employee_name = employee_name

    def mutual_services(self):
        # MUTUAL SERVICES BETWEEN RETAIL AND COMMERCIAL BANKING
        return [
            "Lines of credit",
            "Investment management and accounts",
            "Insurance"
        ]

    def display_services(self):
        pass


# FIRST SUBCLASS
class RetailBanking(ZenithBank):

    def unique_services(self):
        # SERVICES UNIQUE TO RETAIL BANKING
        return [
            "Retirement and education accounts",
            "Loans and mortgages",
            "Checking and saving"
        ]

    def display_services(self):
        # COMBINING MUTUAL AND UNIQUE SERVICES
        all_services = self.mutual_services() + self.unique_services()

        # OUTPUT ON POP-UP
        message = f"Employee: {self.employee_name}\n"
        message += "Division: Retail Banking\n\n"
        message += "Services offered:\n"

        # TO CAPTURE ALL SERVICES
        for service in all_services:
            message += f"  - {service}\n"

        return message


# SECOND SUBCLASS
class GlobalBanking(ZenithBank):

    def unique_services(self):
        # UNIQUE TO GLOBAL BANKING
        return [
            "Multi-currency management services and products",
            "Foreign currency accounts",
            "Foreign currency credit cards",
            "Transborder advisory services",
            "Liquidity management"
        ]

    def display_services(self):
        # GLOBAL BANKING HAS NO MUTUAL SERVICES WITH OTHERS
        all_services = self.unique_services()

        message = f"Employee: {self.employee_name}\n"
        message += "Division: Global Banking\n\n"
        message += "Services offered:\n"

        for service in all_services:
            message += f"  - {service}\n"

        return message


# THIRD AND FINAL SUBCLASS
class CommercialBanking(ZenithBank):

    def unique_services(self):
        # UNIQUE TO COMMERCIAL BANKING
        return [
            "Advisory services"
        ]

    def display_services(self):
        # COMBINATION OF MUTUAL AND UNIQUE SERVICES
        all_services = self.mutual_services() + self.unique_services()

        message = f"Employee: {self.employee_name}\n"
        message += "Division: Commercial Banking\n\n"
        message += "Services offered:\n"

        for service in all_services:
            message += f"  - {service}\n"

        return message


# THE GUI

def show_services():
    # COLLECTING USER INPUT
    name = name_entry.get().strip()
    division = division_entry.get().strip().lower()  

    # DON'T ALLOW USER PROCEED WITHOUT FILLING A NAME
    if not name:
        messagebox.showwarning("Missing Name", "Please enter an employee name.")
        return

    # CREATING AN OBJECT FOR THE DIVISION ENTERED BY THE USER
    if division == "retail":
        employee = RetailBanking(name)

    elif division == "global":
        employee = GlobalBanking(name)

    elif division == "commercial":
        employee = CommercialBanking(name)

    else:
        # ERROR MESSAGE FOR UNMATCHING DIVISION 
        messagebox.showerror("Invalid Division", "Please enter: Retail, Global, or Commercial")
        return

    # FOR DISPLAY
    result = employee.display_services()
    messagebox.showinfo("Zenith Bank Services", result)


# THE WINDOW

# MAIN WINDOW
root = tk.Tk()
root.title("Zenith Bank Nigeria - Employee Services")
root.geometry("400x220")
root.resizable(False, False)
root.configure(bg="WHITE")  # ZENITH BANK RED BACKGROUND

# Title label at the top
title_label = tk.Label(root, text="Zenith Bank Nigeria", font=("Arial", 16, "bold"),
                       bg="white", fg="grey")
title_label.pack(pady=(15, 5))

subtitle_label = tk.Label(root, text="Employee Division Services Lookup",
                          font=("Arial", 10), bg="WHITE", fg="BLACK")
subtitle_label.pack(pady=(0, 15))

# FORM FRAME
form_frame = tk.Frame(root, bg="red")
form_frame.pack()

# EMPLOYEE NAME
name_label = tk.Label(form_frame, text="Employee Name:", bg="grey", fg="white", font=("Arial", 10))
name_label.grid(row=0, column=0, padx=10, pady=6, sticky="e")

name_entry = tk.Entry(form_frame, width=25, font=("Arial", 10))
name_entry.grid(row=0, column=1, padx=10, pady=6)

# DIVISION BOX
division_label = tk.Label(form_frame, text="Division:", bg="grey", fg="white", font=("Arial", 10))
division_label.grid(row=1, column=0, padx=10, pady=6, sticky="e")

division_entry = tk.Entry(form_frame, width=25, font=("Arial", 10))
division_entry.grid(row=1, column=1, padx=10, pady=6)

# EXAMPLER
hint_label = tk.Label(root, text="(Type: Retail, Global, or Commercial)",
                      bg="white", fg="red", font=("Arial", 8), padx=20, )
hint_label.pack()

# SEARCH BUTTON
search_btn = tk.Button(root, text="Show Services", command=show_services,
                       bg="#ffcc00", fg="#003366", font=("Arial", 11, "bold"),
                       padx=10, pady=4, cursor="hand2")
search_btn.pack(pady=12)

# EVENT LOOP, KEEPS THE WINDOW RUNNING AND ALSO GETS INPUTS
root.mainloop()
        
```
