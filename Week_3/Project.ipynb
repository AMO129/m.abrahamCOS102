```python
import tkinter as tk
from tkinter import messagebox, ttk #the ttk is for a drop down widget
import pandas as pd


# Initializing this so that the program knows it exist and is able to read the CSV file
# df = None
#Feeding my program with GIG logistics CSV file

df=pd.read_csv(r"C:\Users\NEW\Desktop\MOSES\PAU SLIDES\YEAR 1\Semester 2\COS102\CURRENT YR\GIG-logistics.csv")


# The main window
root = tk.Tk()
root.title("GIG logistics employee verifier")
root.geometry("400x200")
root.configure(bg="#2c3e50")


#Welcoming user
welcome_label = tk.Label(root, 
                       text="Welcome to GIG Logistics Company\nPlease verify your identity", 
                       fg="white", bg="#2c3e50", 
                       font=("Arial", 12, "bold"))
welcome_label.pack(pady=10)

def verify_employee():

    #User input
    #Accepts both upper and lowercase and also removes accidental space
    user_name = username_entry.get().strip().upper() 
    selected_dept = dept_combobox.get().strip().upper()

    if not user_name or selected_dept == "Choose Department":
        messagebox.showwarning("Input Error, please enter a name and selected a department.")

#Checking if CSV record exist
    match = df[(df['Name'].str.upper() == user_name) & (df['Department'].str.upper() == selected_dept)]

    if not match.empty:
        team_members = df[df['Department'].str.upper() == selected_dept]['Name'].tolist() 
        team_list_str = "\n".join(team_members)

        messagebox.showinfo("Welcome", f"\nWelcome, {user_name}! \nYour team is made up of:\n{team_list_str}")
    else:
        messagebox.showerror(f"Sorry this employee does not exist.")

#Entry box for the username
username_entry = tk.Entry(root, width=23)
username_combobox = ttk.Combobox(root)


# username_combobox.pack(pady=5)
username_entry.pack(pady=15)
username_combobox.set("Username")

# List of available departments
departments = ["Operations", "Finance", "Human Resources", "Technology", "Customer Service", "Sales", "Marketing"]

#Label for the department section
# tk.Label(root, text='Select Department').pack(pady=5)

# The drop-down menu
dept_combobox = ttk.Combobox(root, values=departments, state="readonly", width=20)
dept_combobox.pack()
dept_combobox.set("Choose Department") #THIS LINE SETS A PLACE HOLDER TEXT


# Verifying button
verify_btn = tk.Button(root, 
                       text="Verify Identity", 
                       bg="Blue", 
                       fg="white", 
                       font=("Arial", 10, "bold"), 
                       command=verify_employee)
verify_btn.pack(pady=20)



#To start the application
root.mainloop()
```
