```python
#KEYWORD ARGUMENTS

def printme(str):
    #This prints a passed string
    print(str)
    return
printme(str = 'My name is:') #it's like calling 'x' in a function and assigning '=' a value to 'x' then printing it out
```

```python
def test(name, age):
    #Testing the function even further
    print('Name: ', name)
    print('Age: ', age)
    return
test(age=50, name='Miki')
```

```python
#DEFAULT ARGUMENTS

def test(name, age=35):
    #Test the default function
    print('Name:', name)
    print('Age:', age)
    return
test(age =50, name='Jethro')
test(name='Noro')
```

```python
# VARIABLE ARGUMENTS

def test(arg1, *vartuple):
    #Testing the function
    print('Output is:')
    print(arg1)
    for var in vartuple:
        print(var)
        return
test(10)
test(70,60,50,48)
```

```python
#SCOPE VARIABLE

total=50 #This is global variable
def sum(arg1, arg2):
    #Add both parameters
    total = arg1 + arg2
    print('Inside the function local total: ', total)
    return total
#Now you can call sum function
sum(10,20)
print('Outside the function the total: ', total)
```
