```python
#Practice 1 | cell 1 - Your First Class

class SST:
    prog1 = 'Software Engineering' # class variable

    def __init__(self, name):
        self.name = name #instance variable
    def speak(self):
        print('\nMy name is', self.name)
        print('I study', SST.prog1)

stud1 = SST(input('Enter student 1 name'))
stud2 = SST(input('Enter student 2 name'))

stud1.speak()
stud2.speak()
```

```python
class Addition:
    def __init__(self, f, s):
        self.first = f
        self.second = s
        self.answer = '0.00'
    
    def calculate(self):
        self.answer = self.first + self.second

    def display(self):
        print('\nFirst:', self.first)
        print('Second:', self.second)
        print('\nSum:', self.answer)

num1 = float(input('Enter first number:'))
num2 = float(input('Enter second number:'))

obj = Addition(num1, num2)
obj.display()
obj.calculate()
obj.display()
```
