```python
import tkinter as tk
import random 
# from tkinter import 

#Employees class
class Employee:
    def __init__(self):
        #List of employees
        self.employees = ["Mary Evans", "Eyo Ishan", "Durojaiye Dare", "Adams Ali", "Andrew Ugwu", "Stella Mankinde", "Jane Akibo", 
        "Ago James", "Michell Taiwo", "Abraham Jones" , "Nicole Anide", "Kosi Korso", "Adele Martins", "Emmanuel Ojo", "Ajayi Fatima"]

        #List of tasks
        self.tasks = ["Loading", "Transporting", "Reveiwing Orders", "Customer Service", "Delivering Items"]

    def check_employee(self, name):
        if name.strip().title() in self.employees:
            return True
        else:
            return False

    def get_random_task(self):
        #Picking a random task from the list of tasks
        return random.choice(self.tasks)

#Object for employee system from the Employee class
company_system = Employee()


#The GUI
root = tk.Tk()
root.title("Task Assigner")
root.geometry("400x200")
root.config(bg = "navyblue")

#The submit button functionality

def on_click():
    #Get the text typed in the entry box
    user_name = entry.get()

    #Checking if the user is an employee or not
    if company_system.check_employee(user_name):

        #Assigning tasks
        task = company_system.get_random_task()

        #Configure the label text to welcome the user and show the assigned task
        label.config(text = "Hello " + user_name + " " + "!\nYour task is: " + task)
    else:
        label.config(text = "Sorry, " + user_name + " " + "is not an employee")


#The GUI widgets
# directive text

instr = tk.Label(root, text = "Type your name")
instr.pack(pady=10)

#The name bos
entry = tk.Entry(root)
entry.pack(pady=5)

#Submit button
submit = tk.Button(root, text = "Submit", command = on_click)
submit.pack(pady=10)

#Label for final display
label = tk.Label(root, text = "")
label.pack(pady=10)


#Event(start)
root.mainloop()
```
