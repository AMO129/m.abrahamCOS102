```python
import tkinter as tk


class delivery:
    def __init__(self):
        #Creating a list of locations
        self.locations = ["pau", "epe"]


    def calculate_price(self, location, weight):

        new_location = location.strip().lower()

        #Handling locations out of scope cleanly
        if new_location not in self.locations:
            return "Sorry we only deliver to PAU and Epe"
    
        if new_location == "pau":
            if weight >= 10:
                return "Your delivery fee is N2000"
            else:
                return "Your delivery fee is N1500"
        elif new_location == "epe":
            if weight >= 10:
                return "Your delivery fee is N5000"
            else:
                return "Your delivery fee is N4000"

#Create and Object of the class
my_delivery = delivery()


#The GUI 
root = tk.Tk()
root.title("Delivery Fee Calculator")
root.geometry("350x300")

#The Button Functionality
def verify_price():
    #Getting texts from text boxes
    user_location = location_entry.get()
    weight_text = weight_entry.get()
    
    #Converting weight text to a number in case of mistakes (float handles decimals like 10.5kg)
    try:
        user_weight = float(weight_text)
    except ValueError:
        # If user types words instead of numbers for weight
        result_label.config(text="Please enter a valid number for weight.", fg="red")
        return

    #Using class to get the final price
    final_result = my_delivery.calculate_price(user_location, user_weight)
    
    #Displaying result on the screen
    result_label.config(text=final_result, fg="green")


#GUI Widgets

#Location Input
tk.Label(root, text="Enter Delivery Location (PAU or Epe):").pack(pady=5)
location_entry = tk.Entry(root)
location_entry.pack(pady=5)

#Weight Input
tk.Label(root, text="Enter Package Weight (in kg):").pack(pady=5)
weight_entry = tk.Entry(root)
weight_entry.pack(pady=5)

#Verification Button
verify_button = tk.Button(root, text="Verify Price", command=verify_price)
verify_button.pack(pady=15)

#Result Label
result_label = tk.Label(root, text="", font=("Arial", 10, "bold"))
result_label.pack(pady=10)

#Event(start)
root.mainloop()
```
