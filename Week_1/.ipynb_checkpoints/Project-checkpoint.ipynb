SIMPLE INTEREST

FLOWCHART IMAGE FOR SIMPLE INTEREST



```python
#SIMPLE INTEREST

#COLLECTING USER INPUT

try:
    P = abs(float(input('Enter the principal (P):')))
    R = abs(float(input('Enter the rate (R):')))
    T = abs(float(input('Enter the time in years (T):')))

    
#COMPUTE THE THE AMOUNT(A)

    A = P*(1 + (R/100)*T)

#PRINT/DISPLAY AMOUNT
    print(f'The total simple interest amount is:',A)

#ERROR HANDLING
except ValueError:
    #ERROR MESSAGE 
    print('Error: Please enter figures only')
    
```

COMPOUND INTEREST

FLOWCHART IMAGE FOR COMPOUND INTEREST



```python
#COMPOUND INTEREST

#COLLECTING USER INPUT

try:
    P = abs(float(input('Enter the principal (P):')))
    R = abs(float(input('Enter the rate (R):')))
    n = abs(float(input('Enter times compounded per year (n):')))
    t = abs(float(input('Enter the time in years (t):')))

#COMPUTING COMPOUND INTEREST
    A = P*(1 + R/n)**(n*t)


#PRINT/DISPLAY AMOUNT
    print(f'The total compound interest amount is:',A)

#ERROR HANDLING
except ValueError:
    #ERROR MESSAGE
    print('Error: Please enter figures only')
```

ANNUITY PLAN

FLOWCHART IMAGE FOR ANNUITY PLAN



```python
#ANNUITY PLAN

#COLLECTING USER INPUT

try:
    PMT = abs(float(input('Enter the regular payment amount (PMT):')))
    R = abs(float(input('Enter the rate (R):')))
    n = abs(float(input('Enter the times compounded per year (n):')))
    t = abs(float(input('Enter the time in years (t):')))

#COMPUTING ANNUITY PLAN
    A = PMT * ((1+R/n)**(n*t) - 1)/(R/n)

#PRINT/DISPLAY AMOUNT
    print(f'The total annuity plan amount is:', A)

#ERROR HANDLING
except ValueError:
    #ERROR MESSAGE
    print('Error: Please enter figures only')  
```
