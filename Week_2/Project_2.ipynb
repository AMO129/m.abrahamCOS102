```python
#### Importing the necessary library 

import numpy as np

def solve_polynomial():
    while True:
        print("\n" + "="*45)
        print("WELCOME USER, THIS IS A MINI ROOT CALCULATOR")
        print("="*45)
        print("Select the operation you want to perform:")
        print("1. Quadratic (Ax^2 + Bx + C = 0)")
        print("2. Cubic     (Ax^3 + Bx^2 + Cx + D = 0)")
        print("3. Quartic   (Ax^4 + Bx^3 + Cx^2 + Dx + E = 0)")
        print("4. Exit Program")
        
        choice = input("\nEnter your choice (1-4): ")

        if choice == '4':
            print("Exiting calculator... Goodbye!")
            break
        
        # Mapping choices to the number of coefficients needed; Quadratic needs 3, Cubic needs 4 and Quartic needs 5.
        degree_map = {'1': 3, '2': 4, '3': 5}
        
        labels = ['A', 'B', 'C', 'D', 'E']   #List of coefficients

        if choice in degree_map:
            num_coeffs = degree_map[choice]
            coeffs = []

            #ASSIGNING VALUES TO COEFFICIENTS
            print(f"\nPlease enter the values of the coefficients:")   
            
            try:
                for i in range(num_coeffs):
                    val = float(input(f"  Enter value for {labels[i]}: "))
                    coeffs.append(val)
                
                # A cannot be zero for the chosen operation
                if coeffs[0] == 0:
                    print("\nError: 'A' cannot be 0 because that would reduce the equation's degree; it will not fit the degree of the operation you chose.")
                    continue

                # Calculating roots using NumPy
                roots = np.roots(coeffs)

                print("\n--- RESULTS ---")
                for i, root in enumerate(roots, 1):
                    
                    # Clean up complex numbers for display to avoid long decimal places like 0.00000000001
                    rounded_root = np.round(root, 4)
                    
                    if np.isreal(root):
                        print(f"Root {i} (Real): {rounded_root.real}")
                    else:
                        print(f"Root {i} (Complex): {rounded_root}")
                        
            except ValueError:
                print("\nInvalid input! Please enter numbers only.")
        else:
            print("\nInvalid selection. Please choose 1, 2, 3, or 4.")
            

#To alllow this code/calculator be used universally 
if __name__ == "__main__":
    solve_polynomial()
```
