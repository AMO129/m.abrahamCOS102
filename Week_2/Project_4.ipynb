```python
# Welcome user
print('Welcome to Izifin Technology dear esteemed user.')

try:
    # Collecting user input
    age = abs(int(input('\n Enter your age here: ')))
    exp = abs(float(input('Enter your years of work experience here: ')))
        
        
        #Conditional statement
    if exp > 25 and age >= 55:
            print('Your Annual Tax Revenue(ATR) is: 5,600,000')
        
    elif exp > 20 and age >=45:
            print('Your Annual Tax Revenue(ATR) is: 4,480,000')
        
    elif exp > 10 and age >= 35:
            print('Your Annual Tax Revenue(ATR) is:1,500,000')
        
    else:
            print('Your Annual Tax Revenue(ATR) is:550,000')
            

except ValueError:
        print('Error: Please input a valid age or year(s) of work experience')
```
